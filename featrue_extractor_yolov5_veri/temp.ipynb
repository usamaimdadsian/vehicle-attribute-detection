```python
import pandas as pd
import cv2
import os
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("VeRiAttributesTrainAll.csv")
```

```python
image_folder = "/mnt/linux/stallionSpace/itu/semester3/thesis1/codes/vehicle_color_recognition_old/VeRi"  
```

```python
for i in range(len(df)):
    item = df.iloc[i]

    path = item["path"].replace("/content/VeRi/","")
    x_center, y_center, width, height = item['xcenter'], item['ycenter'], item['width'], item['height']
    image_path = os.path.join(image_folder, path)
    image = cv2.imread(image_path)
    # image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    image_height, image_width, _ = image.shape
    x_center = int(x_center * image_width)
    y_center = int(y_center * image_height)
    width = int(width * image_width)
    height = int(height * image_height)

    # Convert center coordinates to top-left coordinates
    x_min = int(x_center - width / 2)
    y_min = int(y_center - height / 2)

    # Draw rectangle on image
    image = cv2.rectangle(image, (x_min, y_min), (x_min + width, y_min + height), (0, 255, 0), 1)
    target_path =  os.path.join("VeRi_box",path)
    cv2.imwrite(target_path,image)
```
